```python
import pandas as pd

df = pd.read_csv('LOTT/상품분류.txt',sep=',',encoding='cp949')

df.head()
```

```python
print('중분류 수 : ',df['중분류명'].nunique())
print('소분류 수 : ',df['소분류명'].nunique())
```

```python
df.groupby('제휴사')[['제휴사']].count()
```

```python
df.groupby(['제휴사','중분류명'])['소분류명'].count().sort_values(ascending=False)[:30]
```
